### Terminology

1. Relations: In RDBMS, we store data in a tables which contains Rows & Columns. There could be multiple tables which may have a "Relation" & Thats why tables are generally called "Relations" 

2. Attributes: Columns in tables are generally called Attributes

3. Tuples: Rows in tables are called Tuples

4. #### Cardinality of a Relation: No of rows is called "Cardinality of a relation" like if a table has 4 rows then its cardinality of relation will be 4

5. #### Degree of a Relation: No of columns is called "Degree of a relation" like if a table has 5 columns then its degree of a relation will be 5

6. NULL: missing value is called NULL

7. Domain: Specifies what type of values you can store in a column like ENAM can only have data in strings format, Salary can only have data in INT format, Dept column can only have any 1 of 3 values bcoz company only have 3 depts.

### Dbs keys

A key in a dbs is an attribute (column) or set of attributes (set of columns) that uniquely identifies a tuple (row) in a table.



like in above picture roll no & email can become keys as they are unique of every students.


### Different type of keys

#### 1. Super key (Janta)
set of single or combination of columns which uniquely identifies any row is called super key




#### 2. Candidate key (Ummidwaar)
Its the smallest set of attributes (columns) that can be used to uniquely identify a tuple (row) in the table
like rollno & email alone can become candidate key so we dont need other attributes in the set.

#### 3. Primary key (Selected MP)
Any one key which is chosen out of candidate keys is "Primary Key". There can only be 1 primary key.

##### mandatory properties
+ It shouldnt have NULL values - roll no & email
+ It shudbt have duplicate values - roll no & email

##### good to have properties of primary key
+ numerical - roll no
+ small - roll no
+ consistent - roll no


##### Roll No is the best option to become Primary Key



#### 4. Alternate key
Its a candidate key which is not used as primary key like "email"


#### 5. Composite key
Its a primary key which is made up of 2 or more attributes (columns) mostly when no primary keys available



#### 6. Surrogate key
when no attribute (column) to become primary key then we introduce a new attribute to be a primary key & thats called surrogate key




#### 7. Foriegn key
when a primary key from 1 table is used to establish a relationship with another table, its called Foriegn key.


### cardinality of relationship

Type or nature of relationship which can occur b/w 2 entities like name, school etc (column names)

There are only 3 types of possible relationships:


+ 1:1 - person & his driving license: person can only have 1 driving license & 1 driving license can only be associated with 1 person

+ 1:N - student & college branch: 1 student can only be of 1 branch but its opposite is not true, 1 branch can have multiple students

+ M:N - students & courses (on udemy platform): 1 student can be registrd with many courses & 1 courses can have many students


##### How many tables you need while designing dbs:

* If there is a 1:1 relationship b/w 2 entities then you only need 1 table
* If there is a 1:N relationship b/w entities then you need 2 table
* If there is a M:N relationship b/w entities then you need 3 table

### Drawbacks of Dbs

1. Complexity: Managing & Maintaing Dbs can be complex & time consuming.
2. Cost: Its high.
3. Scalability: when amt of data in dbs grows then scalability & speed hampers.
4. Data Integrity: It could be difficult when multiple users are creating & updating data.
5. Security: Increasing threats of cyber attacks is challenging security.
6. Data Migration: Moving data from 1 dbs to another can be difficult.
7. Flexibility: Dbs structue is often rigid & inflexible so you have to plan it carefully at the start of creation
